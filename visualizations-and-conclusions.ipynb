According to the chart above, Thriller, Mystery and Horror movies makes the most return on investment. One of the reasons why these 3 genres have the highest ROI is that the production cost is significantly low. For instance, to film one high grossing action movie you need to spend on average aroun $23-25 millions, which is 9 times bigger than horror movie production cost. Below, we present a table of the mean production budget and return on investment for each of the four selected genres discussed here, as well as a table of the mean return on investment for each genre.

```python
pd.options.display.float_format = '{:.2f}'.format

final_data = pd.read_pickle('analyzed_data/final_data_with_genres.pkl')
horror = final_data[final_data.Horror == 1]
mystery = final_data[final_data.Mystery == 1]
thriller = final_data[final_data.Thriller == 1]
action = final_data[final_data.Action == 1]

horror_df = horror.describe().loc['mean', ['production_budget', 'ROI']]
mystery_df = mystery.describe().loc['mean', ['production_budget', 'ROI']]
thriller_df = thriller.describe().loc['mean', ['production_budget', 'ROI']]
action_df = action.describe().loc['mean', ['production_budget', 'ROI']]

summary_df = pd.DataFrame(asarray([
        horror_df.values, 
        mystery_df.values,
        thriller_df.values,
        action_df.values]).T,
    index = ['Mean Budget ($)', 'Mean ROI (%)'], 
    columns = ['Horror', 'Mystery', 'Thriller', 'Action'])

summary_df
```
